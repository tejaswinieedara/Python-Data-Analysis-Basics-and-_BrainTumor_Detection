#CoinMarketCap


Extracting data from coinmarketcap API

program flow:
* go to CoinMarketCap website
* go to products --> API and obtain API key
* go to documentation to obtain API code
* know your code
* run it in any python environment
* obtain data in the form of json
* normalize the data into a dataframe
* save ii if you need it

```python
 #This example uses Python 2.7 and the python-request library.

from requests import Request, Session
from requests.exceptions import ConnectionError, Timeout, TooManyRedirects
import json

url = 'https://sandbox-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
parameters = {
  'start':'1',
  'limit':'5000',
  'convert':'USD'
}
headers = {
  'Accepts': 'application/json',
  'X-CMC_PRO_API_KEY': 'c28e284b-0912-4aac-ac87-030757fc2e49',
}
session = Session()
session.headers.update(headers)

try:
  response = session.get(url, params=parameters)
  data = json.loads(response.text)
  print(data)
except (ConnectionError, Timeout, TooManyRedirects) as e:
  print(e)
```

```python
import pandas as pd
```

```python
norm=pd.json_normalize(data['data'])
```
