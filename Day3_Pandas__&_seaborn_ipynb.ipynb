LINEAR FIT
* A linear line is rotated to a particular angle untill the distance b/w indvidual points in a graph that are  plotted and the tiltted line is minimum in all cases
* lmplot()

LINEAR REGGRESSION[algorithm]:
* plot the data set
* performing linear fit
* predict the value based on graph
